```python
import pandas as pd
```

Create Dataframe using in-built function read_csv() in pandas

```python
df = pd.read_csv('meddata.csv')
```

```python
df['price'] = df['price'].str.replace(" ","")
```

```python
df['price_per_unit'] = df['price_per_unit'].str.replace(" ","")
```

```python
df
```

In this dataframe we have to fill the values which are not specified

```python
df = df.replace(to_replace ="-",value = 0.0)
```

```python
df.head(60)
```

Transpose dataframe to convert into dictionary so that we can save this data to our django models by using bulk update_or_create 

```python
transpose_df = df.transpose()
```

```python
transpose_df
```

```python
final_df = transpose_df.to_dict()
```

```python
final_df
```
